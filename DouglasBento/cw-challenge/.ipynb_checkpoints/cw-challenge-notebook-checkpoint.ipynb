# Preliminary Findings

## 1. `df_users` Table Insights
- **Missing Data**:
  -  Columns (`score`, `approved_date`, `credit_limit`, `interest_rate`, `loan_term`) have missing values for some users. The `denied_date` column has many null values, likely representing users whose loan applications were never denied.
  - Around **67 users (9% of the total)** are missing key loan data (e.g., score, credit limit), suggesting incomplete credit history.
  
- **Credit Scores**:
  - The average credit score is **5.67**, with a relatively narrow standard deviation of **0.06**, meaning most users have scores within a small range.
  - The minimum credit score is **5.60**, and the maximum is **6.11**. This suggests that the system is serving users with slightly varying credit scores, but all fall into a specific range.

- **Loan Terms & Interest Rates**:
  - The average loan term is approximately **11 months**, with a majority of users receiving loans with terms between **3 to 12 months**.
  - Interest rates range from **2% to 13%**, with an average of **8.6%**.
  
- **Credit Limits**:
  - Credit limits vary significantly, from **$1,000 to $49,800**, with an average of **$13,645**. Users with higher credit scores likely have larger credit limits.

### Preliminary Findings:
- There is some incomplete data in the system for a small portion of users, especially those missing scores and credit limits. This may indicate issues in the approval process or customer segmentation.
- Credit scoring seems tightly grouped around similar scores, which might suggest conservative risk assessment.

## 2. `df_emotional_data` Table Insights
- **Data Size**:
  - This table is quite large, containing **over 2.4 million rows**, reflecting the collection of fine-grained emotional and contextual data for users.
  
- **Intensity and Grade**:
  - The average intensity of emotions is **5.50**, with the range between **1 and 10**. The distribution seems relatively balanced, but further exploration of emotions by users may reveal patterns.
  - The `grade` column (which could represent the "quality" or "impact" of emotional states) has a mean of **4.59**, implying a slight skew towards more positive or neutral emotions.

- **Contextual Variables**:
  - Several categorical columns, like `time_of_day`, `primary_emotion`, `relationship`, and `situation`, will require deeper exploration to understand emotional patterns and their relationships to loan behavior.

### Preliminary Findings:
- The emotional data is dense, with a balanced spread of emotional intensities. The correlation of these emotional intensities with loan approvals or denials may yield interesting insights.
- Further analysis is needed to understand which emotions dominate in certain situations or times of the day.

---

## 3. `df_loans` Table Insights
- **Loan Amounts**:
  - The loan amounts range from **$1,000 to $46,785**, with a mean of **$6,173**. Most loans are small, with 75% of loans below **$6,874**, indicating that the system serves customers with moderate to low loan needs.

- **Installments & Loan Amount Paid**:
  - The average installment amount is **$978**, with substantial variance. Some users may have larger installments due to larger loans or shorter terms.
  - While the maximum amount paid is **$81,090**, the median paid amount is much lower (**$2,247**), suggesting that larger loan amounts are rare.

- **Loan Status**:
  - Out of the 1,188 loans, 241 (20.3%) have missing values for `paid_date`, which may indicate ongoing or defaulted loans. This information will be key to evaluating lending performance.

### Preliminary Findings:
- Most users are borrowing smaller amounts and have relatively low monthly installment amounts.
- The presence of many loans without a `paid_date` suggests some issues with repayment or active loans.

# Data prep

## users table

```python
# Check the percentage of missing values in df_users
df_users.isnull().mean() * 100
```

```python
# Impute missing numerical columns with the median
df_users['score'].fillna(df_users['score'].median(), inplace=True)
df_users['credit_limit'].fillna(df_users['credit_limit'].median(), inplace=True)
df_users['interest_rate'].fillna(df_users['interest_rate'].median(), inplace=True)
df_users['loan_term'].fillna(df_users['loan_term'].median(), inplace=True)

# Impute missing approved_date with a placeholder
df_users['approved_date'].fillna('No Approval', inplace=True)

# Check the updated data
df_users.isnull().sum()
```

```python
df_users['denied_date'].fillna("Not Denied", inplace=True)
```

```python
df_users.isnull().sum()
```

```python
# Create a flag column to indicate denial status
df_users['denial_status'] = df_users['denied_date'].isna().map({True: 'Not Denied', False: 'Denied'})

# Convert denied_date to datetime but retain the denial status separately
df_users['denied_date'] = pd.to_datetime(df_users['denied_date'], errors='coerce')


#convert columns to date format
df_users['approved_date'] = pd.to_datetime(df_users['approved_date'], errors='coerce')
df_users['denied_date'] = pd.to_datetime(df_users['denied_date'], errors='coerce')
```

```python
# Check for duplicates
df_users.duplicated().sum()

# Drop duplicates if any are found
df_users.drop_duplicates(inplace=True)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Ensure that the dates are in datetime format
df_loans['paid_date'] = pd.to_datetime(df_loans['paid_date'])
df_loans['due_date'] = pd.to_datetime(df_loans['due_date'])

# Calculate the difference in months between paid_date and due_date
df_loans['payment_diff_months'] = (df_loans['paid_date'] - df_loans['due_date']).apply(lambda x: x / np.timedelta64(1, 'M'))

# Drop or handle missing values in 'paid_date' (i.e., loans that are not paid yet)
df_loans_filtered = df_loans.dropna(subset=['payment_diff_months'])

# Plot the distribution of the difference
plt.figure(figsize=(10, 6))
plt.hist(df_loans_filtered['payment_diff_months'], bins=20, edgecolor='gray')
plt.title('Distribution of Payment Differences (Months)')
plt.xlabel('Months (Positive: Late, Negative: Advance)')
plt.ylabel('Frequency')
plt.grid(True)
plt.suptitle('Loans', fontsize=16)
plt.show()

# Print summary statistics for payment difference
print(df_loans_filtered['payment_diff_months'].describe())
```

```python
# Map time_of_day to numeric categories
time_of_day_map = {
    'morning': 1,
    'afternoon': 2,
    'evening': 3,
    'night': 4
}
df_emotional_data['time_of_day_numeric'] = df_emotional_data['time_of_day'].map(time_of_day_map)

# Now plot the 3D scatter plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

sc = ax.scatter(df_emotional_data['time_of_day_numeric'], df_emotional_data['intensity'], df_emotional_data['grade'], c=df_emotional_data['grade'], cmap='coolwarm')
ax.set_xlabel('Time of Day (Numeric)')
ax.set_ylabel('Intensity')
ax.set_zlabel('Grade')
plt.title('3D Scatter Plot: Time of Day, Intensity, and Grade')
plt.colorbar(sc)
plt.show()
```

```python
df_emotional_data['time_of_day_numeric'] = df_emotional_data['time_of_day'].map({
    'morning': 1, 'afternoon': 2, 'evening': 3, 'night': 4
})

df_emotional_data['primary_emotion_numeric'] = df_emotional_data['primary_emotion'].astype('category').cat.codes
df_emotional_data['relationship_numeric'] = df_emotional_data['relationship'].astype('category').cat.codes


sns.pairplot(df_emotional_data, hue='intensity', vars=['time_of_day_numeric', 'primary_emotion_numeric', 'grade', 'relationship_numeric'])
plt.title('Pairwise Relationship Analysis (Converted Categorical Variables)', fontsize=16)
plt.show()
```
